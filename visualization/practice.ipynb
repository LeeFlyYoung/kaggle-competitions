```python
import pandas as pd

hotdogs = pd.read_csv('data/hot-dog-contest-winners.csv')
hotdogs.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))  
# 根据 hotdogs 数据集，绘制每年吃热狗数量的柱状图， x 轴为年份，y 轴为吃热狗数量，x 轴标签旋转 45 度，使用 seaborn 库，产生新纪录的年份用红色显示。
ax = sns.barplot(data=hotdogs, x='Year', y='Dogs eaten', hue='New record', dodge=False, palette={True: '#821122', False: '#cccccc'})
# 获取当前 x 轴的所有 ticks 位置
ticks = ax.get_xticks()  # 比如返回 [0, 1, 2, 3, 4, 5, 6, 7]

# 只显示部分 x 轴标签，比如只显示偶数位置的标签
labels = ax.get_xticklabels()  # 获取所有原始标签（比如 'A', 'B', ...）
new_labels = [label.get_text() if i % 4 == 0 else '' for i, label in enumerate(labels)]

# 设置新的 x 轴标签（空字符串代表不显示）
ax.set_xticks(ticks)
ax.set_xticklabels(new_labels)
ax.set_ylabel('Hot dogs and buns (HDB) eaten')
ax.set_title('1980-2010 hotdogs match records', fontsize=16)

# 不显示legend
ax.get_legend().remove()

plt.tight_layout()
# plt.show()
plt.savefig('hotdogs2.pdf', dpi=300)
```

```python
import pandas as pd
unemployment = pd.read_csv('data/unemployment-rate-1948-2010.csv')
unemployment.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
ax = sns.regplot(data=unemployment, x=unemployment.index, y='Value', line_kws= {'color':'black'},scatter_kws={'color':'none', 'edgecolor': 'black'} )
```

```python
import seaborn as sns
from statsmodels.nonparametric.smoothers_lowess import lowess
import matplotlib.pyplot as plt

sns.scatterplot(data=unemployment, x=unemployment.index, y='Value', color='none', edgecolor='black', s=20)
loess_smoothed = lowess(unemployment.Value, unemployment.index, frac=0.3)  # frac 控制平滑程度
x_loess = loess_smoothed[:, 0]
y_loess = loess_smoothed[:, 1]

plt.plot(x_loess, y_loess, color='black', linewidth=2, label='LOESS 平滑曲线')
```

```python
import pandas as pd
fields = pd.read_csv('data/field_of_interest.csv', header = None, names=['Field', 'Count'])
fields = fields[:: -1]
fields.head()
```

```python
import pandas as pd
obama = pd.read_csv('data/obama.csv')
obama.head()
```

```python
# 绘制堆叠柱状图，呈现不同问题的支持、反对、不发表意见比例, 柱状图上不同类别增加数字标签
import matplotlib.pyplot as plt
# 设置中文字体，使用系统支持中文的字体，比如 "SimHei"（黑体）、"Microsoft YaHei"（微软雅黑）
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 指定默认字体（黑体）, 可以换成 'Microsoft YaHei', 'SimSun' 等
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题   

ax = obama.plot(kind='bar', rot=45 ,x='问题', stacked=True, figsize=(10,6), color=["#59A3ED", "#A3E1F9", '#cccccc'])




for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.text(x + width/2, 
            y + height/2, 
            '{:.0f}'.format(height), 
            horizontalalignment='center', 
            verticalalignment='center')
plt.show()


```

```python
import pandas as pd
crime = pd.read_csv('data/crimeRatesByState2005.csv')
crime.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(crime,x = 'murder', y= 'burglary')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.nonparametric.smoothers_lowess import lowess

crime2 = crime.loc[~crime['state'].isin(['District of Columbia', 'United States'])]
sns.set_style('whitegrid')
ax = sns.scatterplot(crime2, x = 'murder', y = 'burglary', color='#cccccc')
ax.set_xlim(0, 12)
ax.set_ylim(0, 1300)
ax.spines[:].set_visible(False)
loess_smoothed = lowess(crime2.burglary, crime2.murder, frac=0.5)  # frac 控制平滑程度
x_loess = loess_smoothed[:, 0]
y_loess = loess_smoothed[:, 1]

plt.plot(x_loess, y_loess, color='black', linewidth=2, label='LOESS 平滑曲线')

plt.show()
```

```python
def my_plot( x, y, color):
    sns.scatterplot(x= x, y = y, color = "#cccccc")
    loess_smoothed = lowess(y, x, frac=0.5)  # frac 控制平滑程度
    x_loess = loess_smoothed[:, 0]
    y_loess = loess_smoothed[:, 1]

    plt.plot(x_loess, y_loess, color='red', linewidth=1)
    
sns.PairGrid(crime2.drop('population', axis = 1) ).map(my_plot)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=crime2, x='murder', y = 'burglary', size='population', sizes = (100, 4000), color='#cccccc', alpha=0.5)
for i, txt in enumerate(crime2.state):
    plt.annotate(txt, (crime2['murder'].iloc[i], crime2['burglary'].iloc[i]), fontsize=9, ha='center')
plt.tight_layout()
plt.show()
```
